## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

print("Libraries imported successfully!")
```

## Step 1: Data Loading
Load dataset dari folder `namadataset_raw/`

```python
# Load data
# Sesuaikan path dengan nama file dataset Anda
# Contoh: df = pd.read_csv('../namadataset_raw/data.csv')

# Jika menggunakan sample data:
np.random.seed(42)
sample_data = {
    'feature_1': np.random.randn(100),
    'feature_2': np.random.randn(100),
    'feature_3': np.random.choice(['A', 'B', 'C'], 100),
    'target': np.random.choice([0, 1], 100)
}
df = pd.DataFrame(sample_data)

print("Data loaded successfully!")
print(f"Shape: {df.shape}")
print(f"\nFirst few rows:")
df.head()
```

## Step 2: Exploratory Data Analysis (EDA)

```python
# Dataset Info
print("Dataset Information:")
print(f"Shape: {df.shape}")
print(f"\nColumn Names and Types:")
print(df.dtypes)
print(f"\nBasic Statistics:")
df.describe()
```

```python
# Check for missing values
print("Missing Values:")
missing = df.isnull().sum()
print(missing)
print(f"\nTotal missing: {missing.sum()}")
```

```python
# Check for duplicate rows
print(f"Duplicate rows: {df.duplicated().sum()}")
print(f"Unique rows: {len(df)}")
```

```python
# Visualize distributions
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

numeric_cols = df.select_dtypes(include=[np.number]).columns
for idx, col in enumerate(numeric_cols):
    ax = axes[idx // 2, idx % 2]
    df[col].hist(ax=ax, bins=20)
    ax.set_title(f'Distribution of {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
print("EDA visualization completed!")
```

## Step 3: Data Preprocessing

```python
# 3a. Handle Missing Values
print("Step 3a: Handling Missing Values")
df_processed = df.dropna()
print(f"Rows after removing missing values: {len(df_processed)}")
```

```python
# 3b. Remove Duplicates
print("Step 3b: Removing Duplicates")
before = len(df_processed)
df_processed = df_processed.drop_duplicates()
after = len(df_processed)
print(f"Rows removed: {before - after}")
print(f"Rows after removing duplicates: {after}")
```

```python
# 3c. Data Type Conversion
print("Step 3c: Data Type Conversion")
# Contoh: konversi string ke kategori
for col in df_processed.select_dtypes(include=['object']).columns:
    unique_values = df_processed[col].nunique()
    if unique_values < 10:
        df_processed[col] = df_processed[col].astype('category')
        print(f"Converted {col} to category type")

print(f"\nData types after conversion:")
print(df_processed.dtypes)
```

```python
# 3d. Normalization/Scaling
print("Step 3d: Normalization/Scaling")
numeric_cols = df_processed.select_dtypes(include=[np.number]).columns.tolist()
scaler = StandardScaler()
df_processed[numeric_cols] = scaler.fit_transform(df_processed[numeric_cols])
print(f"Scaled {len(numeric_cols)} numeric columns")
print(f"\nData after scaling:")
df_processed.describe()
```

## Step 4: Data Validation

```python
# Validate preprocessed data
print("Data Validation Results:")
print(f"Original shape: {df.shape}")
print(f"Processed shape: {df_processed.shape}")
print(f"\nMissing values: {df_processed.isnull().sum().sum()}")
print(f"Duplicate rows: {df_processed.duplicated().sum()}")
print(f"\nData types:")
print(df_processed.dtypes)
print(f"\n✓ Data is ready for model training!")
```
